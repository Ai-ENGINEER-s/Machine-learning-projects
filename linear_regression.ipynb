```python
#  let's try some numbers for linear regression function 

w = 100
b = 100 
print(f"w : {w}")
print(f"w : {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0] 
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```


### 1. La pente (`w`)

La **pente** (repr√©sent√©e par `w`) est un terme qui vient du concept de la droite dans un graphique. 

- Imagine que tu dessines une ligne droite sur un graphique. La pente de cette ligne te dit √† quel point la ligne est "inclin√©e" ou "raide".
  
- **Exemple concret** : Imagine que tu montes une colline. Si la colline est tr√®s raide, c'est comme si la pente de la droite est grande. Si la colline est presque plate, la pente est faible.

### 2. L'ordonn√©e √† l'origine (`b`)

L'**ordonn√©e √† l'origine** (repr√©sent√©e par `b`) est un autre terme li√© aux graphiques.

- C'est le point o√π la ligne traverse l'axe vertical (l'axe des `y` sur un graphique). En d'autres termes, c'est la hauteur √† laquelle la ligne commence si on regarde l√† o√π elle coupe l'axe des `y`.

- **Exemple concret** : Imagine que tu te tiens debout au bord d'une piscine, au niveau de l'eau. Si tu descends dans la piscine, le niveau de l'eau est plus bas (c'est comme si `b` √©tait n√©gatif). Si tu montes sur un plongeoir, tu es plus haut (c'est comme si `b` √©tait positif).

### 3. La sortie (`f(x)` ou `f_wb`)

La **sortie** (aussi appel√©e la pr√©diction) est le r√©sultat que donne le mod√®le apr√®s avoir fait le calcul avec `x`, `w`, et `b`.

- **Exemple concret** : Imagine que tu utilises une r√®gle de trois pour calculer combien d'argent tu aurais si tu √©changeais des euros en dollars. La sortie serait le montant en dollars apr√®s avoir fait le calcul.

### Mettons tout √ßa ensemble

Quand on parle d'un **mod√®le lin√©aire**, on parle d'une √©quation qui ressemble √† √ßa :

\[
f(x) = w \times x + b
\]

- **`x`** : C'est l'entr√©e, ce qu'on met dans le mod√®le. Par exemple, si on essaie de pr√©dire la taille d'une plante en fonction du temps, `x` pourrait √™tre le nombre de jours.
- **`w`** : C'est la pente, qui dit √† quel point la taille de la plante change pour chaque jour qui passe.
- **`b`** : C'est l'ordonn√©e √† l'origine, qui pourrait repr√©senter la taille de la plante au jour 0 (au tout d√©but).
- **`f(x)`** : C'est la sortie, la taille pr√©dite de la plante apr√®s un certain nombre de jours.

### Exemple pratique

Supposons que tu as un mod√®le qui pr√©dit combien d'argent tu gagnes apr√®s avoir travaill√© `x` heures. Si le mod√®le dit :

\[
f(x) = 10 \times x + 50
\]

- **`w = 10`** : Cela signifie que pour chaque heure de travail (chaque augmentation de `x`), tu gagnes 10 ‚Ç¨ de plus.
- **`b = 50`** : C'est l'argent que tu gagnes m√™me si tu ne travailles pas (par exemple, un bonus fixe de 50 ‚Ç¨).
- **`x = 5`** : Si tu travailles 5 heures, la sortie serait :

\[
f(5) = 10 \times 5 + 50 = 100 ‚Ç¨
\]

Donc, apr√®s 5 heures de travail, tu gagnerais 100 ‚Ç¨ selon ce mod√®le.

```python

tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

# x_train is the input variable 
# y_train is the target variable (price in 1000s of dollars )
x_train = np.array([1, 2])
y_train = np.array([300, 500])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")

# Let's calculate the number of training examples
m = x_train.shape[0]
print(f"Number of training examples is : {m}")

# Training example x_i, y_i 
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"x^({i}),y^({i}) = ({x_i}, {y_i})")

# Plotting the data
plt.scatter(x_train, y_train , marker='x', c='red', s=100)

# Define w and b as scalars (not tuples)
w = 100  # Un nombre
b = 100  # Un nombre
print(f"w : {w}")
print(f"b : {b}")

# Define the compute_model_output function
def compute_model_output(x, w, b):
    m = x.shape[0] 
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb

# Compute the model output
tmp_f_wb = compute_model_output(x_train, w, b)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
